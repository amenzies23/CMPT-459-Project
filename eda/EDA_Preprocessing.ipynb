Here are helper functions that can help us explore the dataset.

The `get_features` function removes non-predictive or categorical columns from the dataset to keep only the numerical features used for model training/data exploration.

```python
# Extract Features for Model
def get_features(df, features = ["Plant_ID","Plant_Health_Status"]):
    return df.drop(columns = features)
```

The `set_types` function converts the Timestamp column to a proper datetime format, rounds it to the nearest hour.

```python
def set_types(df):
    df['Timestamp'] = pd.to_datetime(df['Timestamp']).dt.floor("h")
    return df.set_index("Timestamp")
```

The `standardize` function scales all numerical features in the DataFrame using standardization, which transforms each feature to have a mean of 0 and a standard deviation of 1. This ensures that all features contribute equally to the model and prevents features with larger numeric ranges from dominating.

```python
def standardize(df, index = None):
    x = StandardScaler().fit_transform(df)
    return pd.DataFrame(data=x, columns=df.columns, index=index)
```

The `extract_feature` function chains the dataset preprocessing steps (pipeline): 
- formats timestamps
- removes unnecessary columns
- applies feature standardization (scaling numerical values).
- etc.

```python
def extract_features(df):
    return (df.pipe(set_types).pipe(get_features).pipe(standardize, df['Timestamp']))
```

Finally, we execute the feature extraction pipeline.

```python
X_scaled = extract_features(data)
```

# Data Distribution Visualization
First, we wanted to visualize the distribution of the raw dataset. Since we are working with a time series, we wanted to visualize this in a way to see any trends of data values over time. To begin, we potted box plots of the distribution of plant values per day for each feature. This also includes a mean trend line to get an idea of the mean values across plants throughout the full month period.

```python
feature_columns = get_features(data).columns.tolist()
feature_columns.remove("Timestamp")
```

```python
def plot_daily_boxes_all_plants(df, feature):
    data = df.copy()
    data["Timestamp"] = pd.to_datetime(data["Timestamp"], errors="coerce").dt.floor("s")
    # Bucket to days
    data["Day"] = data["Timestamp"].dt.normalize()
    day_index = np.sort(data["Day"].unique())
    box_data = [data.loc[data["Day"].eq(day), feature].to_numpy() for day in day_index]

    positions = mdates.date2num(day_index)
    width = 0.5 
    fig_w = max(14.0, len(day_index) * 0.5)
    fig, ax = plt.subplots(figsize=(fig_w, 5))
    box_plot = ax.boxplot(
        box_data,
        positions=positions,
        widths=width,
        manage_ticks=False,
        showfliers=False,
        patch_artist=True,
    )
    for box in box_plot["boxes"]:
        box.set(facecolor="#cfe2ff", edgecolor="#2554a3", alpha=0.7, linewidth=1.0)
    for med in box_plot["medians"]: med.set(color="#000", linewidth=1.6)
    for whisk in box_plot["whiskers"]: whisk.set(color="#2554a3", linewidth=1.0)
    for cap in box_plot["caps"]: cap.set(color="#2554a3", linewidth=1.0)

    # Mean line across days
    means = [np.nanmean(v) if len(v) else np.nan for v in box_data]
    ax.plot(day_index, means, linewidth=2.0, alpha=0.95)
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
    fig.autofmt_xdate(rotation=45)
    ax.grid(True, alpha=0.3)
    ax.set_title(f"{feature} â€” daily boxes across all plants + mean")
    ax.set_xlabel("Date")
    ax.set_ylabel(feature)
    plt.tight_layout()
    plt.show()
```

```python
for f in feature_columns:
    plot_daily_boxes_all_plants(data, f)
```

## Label Distribution

In our Dataset we have 3 classes. Healthy, Moderately Stressed and Highly Stressed indicating the stress level for each plant.  The Distribution of Classes is fairly even, however healthy plants make up 25% of the examples where stressed plant make up 75% of examples

```python
y = data.loc[:,['Timestamp','Plant_Health_Status']].pipe(set_types)
y.index = y.index.floor("h")
y.head()
```

```python
proportions = y.value_counts(normalize = True)
plt.figure(figsize = (8,4))
sns.barplot(proportions.reset_index(),x = "Plant_Health_Status",y='proportion')
plt.title("Label Distribution of Plant Dataset")
_ = 0 # not show plotting info
```

Since our dataset is a timeseries dataset, we wanted to explore whether our labels had a time dependence. We computed the proportion of classes at different timescales.  By normalizing them we treat them as probabilities for each class.

```python
def proportion(group):
    group = group.drop(columns = 'Timestamp')
    return group.value_counts(normalize = True)
```

```python
stress_levels = y.reset_index().groupby("Timestamp").apply(proportion).reset_index()
stress_per_sample = stress_levels.pivot(
    index='Timestamp',
    columns='Plant_Health_Status',
    values='proportion'
).fillna(0)
```

```python
order = ["Healthy","Moderate Stress", "High Stress"]
daily = stress_per_sample.resample("D").mean()[order]
weekly = stress_per_sample.resample("W").mean()[order]
```

## Label Probabilities over Time
Over the month that our dataset is sampled for we do see periodic changes in the probability of each class.  This suggests that our features do have a time dependence that could obscure informative changes in the features. This makes sense as plants absorb moisture and nutrients gradually over weeks or months, not within a single day.

```python
daily.plot(kind='bar', stacked=True, colormap='summer', figsize=(8,6))
plt.legend(frameon = False,bbox_to_anchor = (1.1,1))
plt.title("Hourly Stress Distribution")
plt.grid(False)
```

Our data is sampled 4 times per day. 
we show below for each sampling time the probability of each class.
We see that the probability does not change over the course of the day. This reaffirms our assumption previously that plants do not absorb all moisture/nurients within a day.

```python
stress_per_sample['hour'] = stress_per_sample.index.hour
hourly = stress_per_sample.groupby('hour').mean()[order]
hourly.plot(kind='bar', stacked=True, colormap='summer', figsize=(8,6))
plt.legend(frameon = False,bbox_to_anchor = (1.1,1))
plt.title("Probability for each Label at Daily Sampling Points")
plt.xlabel("Sampling Time")
plt.grid(axis = 'x')
```

## Correlation / Feature Extraction
Since our data is a time series we extract wavelet features using the stationary wavelet transform.  This allow us to add features that correspond to different frequency levels of the features and incorperate the time dependencies of the dataset.

```python
def wavelet(col, level = 3, method = 'db2'):
    '''
        Compute Stationary Wavelets for A Given Column
        Params: 
        * col: pd.Series column to perform DWT on
        * level: int number of resolution to look at
        * method: str method to use
    '''
    waves = pwt.swt(col, wavelet = method, level = level)
    cols = pd.DataFrame(index=col.index)
    for l in range(level):
        cols[f'{col.name}_a_{l}'] = waves[l][0]
        cols[f'{col.name}_d_{l}'] = waves[l][0]
    return cols
```

```python
def wavelet_features(df, levels = 3, method = 'db2'):
    '''
        Applies wavelet function column wise.  Output is axpanded into a larger dataframe
        * Pandas Dataframe with all numeric columns
        * level: int number of resolution to look at
        * method: str method to use
    '''
    wave_dfs = [wavelet(df[col],level=levels, method=method) for col in df.columns]
    return pd.concat(wave_dfs, axis=1)
```

We will improve that with PCA, extracting orthogonal bases from our data and reducing the dimensionality. 

## Dimensionality Reduction (PCA)
Looking at the explained variance ratio, allows us to see how many PCs we need to keep to explain a certain percentage of the variance of our dataset.  From the plot below we only need 10 PCs for our dataset.

```python
pca = PCA()
transformed = pca.fit_transform(w)
sns.lineplot(pca.explained_variance_ratio_)
plt.xlabel("Number of PCs")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA Cumulative Explained Variance")
```

```python
columns = [f"PC{i+1}" for i in range(10)]
extracted = pd.DataFrame(index= w.index, columns = columns, data = transformed[:,:10])
extracted.head()
```
